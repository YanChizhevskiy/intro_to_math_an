1. Даны три множества a,b и с. Необходимо выполнить все изученные виды бинарных операций над всеми комбинациями множеств.

```python
import itertools
```

```python
a = {1, 2, 3, 10}
b = {2, 3, 4, 20}
c = {3, 4, 5, 10, 20}
```

```python
print('Пересечение а и b, b и a: ', set(a).intersection(set(b)))
print('Пересечение а и с, c и a: ', set(a).intersection(set(c)))
print('Пересечение b и с, c и b: ', set(b).intersection(set(c)))
print('Пересечение а, b, с: ', set(a).intersection(set(b).intersection(set(c))))

print('Объединение а и b, b и a: ', set(a).union(set(b)))
print('Объединение а и с, c и a: ', set(a).union(set(c)))
print('Объединение b и с, c и b: ', set(b).union(set(c)))
print('Объединение a, b, с: ', set(a).union(set(b).union(set(c))))

print('Разность а и b: ', set(a).difference(set(b)))
print('Разность b и a: ', set(b).difference(set(a)))
print('Разность а и c: ', set(a).difference(set(c)))
print('Разность c и a: ', set(c).difference(set(a)))
print('Разность b и c: ', set(b).difference(set(c)))
print('Разность c и b: ', set(c).difference(set(b)))
print('Разность a, b, c: ', set(a).difference(set(b)).difference(set(c)))
print('Разность b, c, a: ', set(b).difference(set(c)).difference(set(a)))
print('Разность c, a, b: ', set(c).difference(set(a)).difference(set(b)))

print('Симметрическая разность а и b, b и a: ', set(a).symmetric_difference(set(b)))
print('Симметрическая разность а и с, с и a: ', set(a).symmetric_difference(set(c)))
print('Симметрическая разность b и с, с и b: ', set(b).symmetric_difference(set(c)))
print('Симметрическая разность a,b,c: ', set(a).symmetric_difference(set(b)).symmetric_difference(set(c)))

print('Декартово произведение а и b: ', [(x, y) for x, y in itertools.product(a, b)])
print('Декартово произведение b и a: ', [(x, y) for x, y in itertools.product(b, a)])
print('Декартово произведение а и c: ', [(x, y) for x, y in itertools.product(a, c)])
print('Декартово произведение c и a: ', [(x, y) for x, y in itertools.product(c, a)])
print('Декартово произведение b и c: ', [(x, y) for x, y in itertools.product(b, c)])
print('Декартово произведение c и b: ', [(x, y) for x, y in itertools.product(c, b)])
print('Декартово произведение a, b, c: ', [(x, y, z) for x, y, z in itertools.product(a, b, c)])
print('Декартово произведение b, c, a: ', [(x, y, z) for x, y, z in itertools.product(b, c, a)])
print('Декартово произведение c, a, b: ', [(x, y, z) for x, y, z in itertools.product(c, a, b)])
```
